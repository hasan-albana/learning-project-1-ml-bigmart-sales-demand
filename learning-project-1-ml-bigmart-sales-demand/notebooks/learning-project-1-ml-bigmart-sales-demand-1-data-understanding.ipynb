| 10 | Outlet_Location_Type          | Tier level of the outlet location |

### Item Related Features

```python
df.info()
```

```python
item_related_features = [
    "Item_Fat_Content", "Item_Type", "Item_MRP", "Item_Visibility"
]

for i in item_related_features:
    if df[i].dtype == 'object':
        plt.figure(figsize=(12, 6))
        sns.boxplot(data=df, x=i, y="Item_Outlet_Sales")
        if df[i].nunique() > 10:
            plt.xticks(rotation=90)
        else:
            plt.xticks(rotation=0)
        plt.show()

    elif df[i].dtype in ['int64', 'float64']:
        plt.figure(figsize=(12, 6))
        sns.histplot(data=df, x=i, y="Item_Outlet_Sales")
        plt.show()
```

### Outlet Related Features

```python
outlet_related_features = [
    "Outlet_Establishment_Year",
    "Outlet_Size", "Outlet_Location_Type", "Outlet_Type"
]
for i in outlet_related_features:
    sns.boxplot(data=df, x=i, y="Item_Outlet_Sales")
    if i == "Outlet_Establishment_Year" or i == "Outlet_Type":
        plt.xticks(rotation=45)
    else:
        plt.xticks(rotation=0)
    plt.show()
```
